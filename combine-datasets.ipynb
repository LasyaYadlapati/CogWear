```python
pilot_baseline_temp = pd.DataFrame()
for i in range(0, 11):
    df = pd.read_csv(f'data/pilot/{i}/baseline/empatica_temp.csv')
    df['time'] = pd.to_datetime(df['time'], unit='s')
    df.insert(0, 'subject', i)
    df['time_elapsed'] = (df['time'] - df['time'].iloc[0]).dt.total_seconds()
    df['time'] = df['time'].dt.time
    pilot_baseline_temp = pd.concat([pilot_baseline_temp, df])
pilot_baseline_temp = pilot_baseline_temp.reset_index(drop=True)
```

```python
# pre
final_pre = pd.DataFrame()
df_dict = dict()

for i in range(11, 24):
    for metric in ['bvp', 'temp', 'eda']:
        if i != 19:
            df = pd.read_csv(f'data/survey_gamification/{i}/pre/survey/questionnaires.csv')
            df.insert(0, 'subject', i)
            df['TimeSpentSeconds'] = pd.to_timedelta(df['TimeSpent']).dt.total_seconds()
            df = df.drop(columns=['Unnamed: 0', 'TimeSpent'])
            df = df[df['QuestionnaireName'] != 'NFR']
            
            # check if gamified or non-gamified
            if df['Gamified'].iloc[0] == 'yes':
                df_name = f"gamified_{metric}"
                gamified = pd.read_csv(f'data/survey_gamification/{i}/pre/survey/empatica_{metric}.csv')
                gamified.insert(0, 'subject', i)
                gamified['time'] = pd.to_datetime(gamified['time'], unit='s')
                gamified['time_elapsed'] = (gamified['time'] - gamified['time'].iloc[0]).dt.total_seconds()
                gamified = gamified.drop(columns=['time'])
                
                # initialize key if it doesn't exist, otherwise append
                if df_name in df_dict:
                    df_dict[df_name] = pd.concat([df_dict[df_name], gamified])
                else:
                    df_dict[df_name] = gamified
            else:
                df_name = f"non_gamified_{metric}"
                non_gamified = pd.read_csv(f'data/survey_gamification/{i}/pre/survey/empatica_{metric}.csv')
                non_gamified.insert(0, 'subject', i)
                non_gamified['time'] = pd.to_datetime(non_gamified['time'], unit='s')
                non_gamified['time_elapsed'] = (non_gamified['time'] - non_gamified['time'].iloc[0]).dt.total_seconds()
                non_gamified = non_gamified.drop(columns=['time'])
                
                # initialize key if it doesn't exist, otherwise append
                if df_name in df_dict:
                    df_dict[df_name] = pd.concat([df_dict[df_name], non_gamified])
                else:
                    df_dict[df_name] = non_gamified

    # append current df to the final df
    final_pre = pd.concat([final_pre, df]).reset_index(drop=True)

# post
final_post = pd.DataFrame()
for i in range(11, 24):
    for metric in ['bvp', 'temp', 'eda']:
        if i not in [19, 20, 21]:
            df = pd.read_csv(f'data/survey_gamification/{i}/post/survey/questionnaires.csv')
            df.insert(0, 'subject', i)
            df['TimeSpentSeconds'] = pd.to_timedelta(df['TimeSpent']).dt.total_seconds()
            df = df.drop(columns=['Unnamed: 0', 'TimeSpent'])
            df = df[df['QuestionnaireName'] != 'NFR']
            
            # check if gamified or non-gamified
            if df['Gamified'].iloc[0] == 'yes':
                df_name = f"gamified_{metric}"
                gamified = pd.read_csv(f'data/survey_gamification/{i}/post/survey/empatica_{metric}.csv')
                gamified.insert(0, 'subject', i)
                gamified['time'] = pd.to_datetime(gamified['time'], unit='s')
                gamified['time_elapsed'] = (gamified['time'] - gamified['time'].iloc[0]).dt.total_seconds()
                gamified = gamified.drop(columns=['time'])
                
                # initialize key if it doesn't exist, otherwise append
                if df_name in df_dict:
                    df_dict[df_name] = pd.concat([df_dict[df_name], gamified], ignore_index=True)
                else:
                    df_dict[df_name] = gamified

            else:
                df_name = f"non_gamified_{metric}"
                non_gamified = pd.read_csv(f'data/survey_gamification/{i}/post/survey/empatica_{metric}.csv')
                non_gamified.insert(0, 'subject', i)
                non_gamified['time'] = pd.to_datetime(non_gamified['time'], unit='s')
                non_gamified['time_elapsed'] = (non_gamified['time'] - non_gamified['time'].iloc[0]).dt.total_seconds()
                non_gamified = non_gamified.drop(columns=['time'])
                
                # initialize key if it doesn't exist, otherwise append
                if df_name in df_dict:
                    df_dict[df_name] = pd.concat([df_dict[df_name], non_gamified], ignore_index=True)
                else:
                    df_dict[df_name] = non_gamified

    # append current df to the final df
    final_post = pd.concat([final_post, df], ignore_index=True)
```

```python
df_dict['gamified_bvp'].to_csv('./cleaned_data/survey/gamified_bvp.csv')
df_dict['non_gamified_bvp'].to_csv('./cleaned_data/survey/non_gamified_bvp.csv')
df_dict['gamified_temp'].to_csv('./cleaned_data/survey/gamified_temp.csv')
df_dict['non_gamified_temp'].to_csv('./cleaned_data/survey/non_gamified_temp.csv')
df_dict['gamified_eda'].to_csv('./cleaned_data/survey/gamified_eda.csv')
df_dict['non_gamified_eda'].to_csv('./cleaned_data/survey/non_gamified_eda.csv')
```
